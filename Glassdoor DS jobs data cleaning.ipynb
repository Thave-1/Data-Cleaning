## Glassdoor DS Jobs Data Cleaning
<b>Task</b>
- Drop <b>Index</b> column (done)
- Clean <b>Salary Estimate</b> column
`Seperate the text 'Glassdoor est' from the numbers` and
`Extract the minimum and maximum estimate to form new columns`(done)
- Clean the <b>Company Name</b> column (done)
- In the <b>Location</b> column, `seperate the States from the country`
- Drop the <b>Job Description</b> column (done)
- Adjust the column names

```python
import pandas as pd
import numpy as np
```

```python
jobs = pd.read_csv('files/Uncleaned_DS_jobs.csv', encoding ='utf-8')
jobs.head(3)
```

```python
jobs.drop('index',axis=1,inplace=True)
```

```python
jobs.columns
```

```python
jobs.shape
```

```python
jobs.info()
```

```python
jobs['Salary Estimate'].unique()
```

```python
pattern = '([$\w]*K\-[$\w]*K)'
jobs['Salary Estimate'] = jobs['Salary Estimate'].str.extract(pattern)[0]
jobs.head()
```

```python
pattern = '(\$)([\w]*)(K)(\-)(\$)([\w]*)(K)'
jobs['Min_Salary_Estimate'] = jobs['Salary Estimate'].str.extract(pattern)[1].astype('int') * 1000
jobs['Max_Salary_Estimate'] = jobs['Salary Estimate'].str.extract(pattern)[5].astype('int') * 1000
```

```python
jobs.drop('Salary Estimate', axis=1, inplace=True)
```

```python
pattern = '([-\w ]*)'
jobs['Company Name'] = jobs['Company Name'].str.extract(pattern)[0]
```

```python
pattern = '([-\w ]*)(, )(\w{2})'
jobs['City'] = jobs['Location'].str.extract(pattern)[0]
jobs['State'] = jobs['Location'].str.extract(pattern)[2]
```

```python
jobs.drop('Job Description',axis=1, inplace=True) # Drop JOB DESCRIPTION column
```

```python
jobs.columns
```

```python
jobs.rename(columns={'Job Title':'Job_Title',
             'Company Name':'Company_Name',
             'Type of ownership':'Type_of_ownership'},inplace=True)
```

```python
jobs.head()
```

```python
# Convert the -1 values to NaN
jobs.replace({-1:np.nan,
             '-1':np.nan}, 
             inplace=True) 
```
